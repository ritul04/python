```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
```

```python
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
table=soup.find("table",class_="table")
```

```python
rows = table.tbody.find_all("tr",limit=10)

data = []
for row in rows:
    cells = row.find_all("td")
    
    # Extract the data from each cell
    rank = cells[0].text.strip()
    team = cells[1].text.split('\n')[-3]
    matches = cells[2].text.strip()
    points = cells[3].text.strip()
    rating = cells[4].text.strip()
    
    
    data.append((rank, team, matches, points, rating))

df = pd.DataFrame(data, columns=["Rank", "Team", "Matches", "Points", "Rating"])
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

# Get the table with the rankings
table = soup.find("table", attrs={"class": "table"})

# Get the rows of the table
rows = table.tbody.find_all("tr")

# Initialize the list to store the rankings
rankings = []

# Loop through each row
for row in rows[:10]:
    # Get the player name, team, and rating
    player_name = row.find("a").text
    team = row.find("td", attrs={"class": "table-team"}).text
    rating = row.find("td", attrs={"class": "table-rating"}).text
    
    # Add the information to the list
    rankings.append({"Player": player_name, "Team": team, "Rating": rating})

# Create the dataframe
df = pd.DataFrame(rankings)

# Print the data
df
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
table = soup.find("table", class_="table")
rows = table.tbody.find_all("tr")

data = []
for row in rows[:10]:
    player = row.find("td", class_="table-body__cell name").text.strip()
    team = row.find("td", class_="table-body__cell flag").find("span").text.strip()
    rating = int(row.find("td", class_="table-body__cell rating").text.strip())
    data.append([player, team, rating])

df = pd.DataFrame(data, columns=["Player", "Team", "Rating"])
print(df)
```

# ANSWER 6 -a

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Scraping the data
url = "https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
res = requests.get(url)

soup = BeautifulSoup(res.text, "html.parser")
table = soup.find("table", {"class": "table"})

# Getting the headers
header_row = table.find("thead").find("tr")
header_cols = header_row.find_all("th")
headers = [col.text.strip() for col in header_cols]

# Getting the rows
body_rows = table.find("tbody").find_all("tr")
data = []
for row in body_rows:
    cols = row.find_all("td")
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Creating the dataframe
df = pd.DataFrame(data, columns=headers)

# Filter only top 10 ODI teams in women's cricket
df = df.iloc[:10, :]

# Display the dataframe
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.icc-cricket.com/rankings/womens/odi"

# Make a GET request to the website
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the rankings information
table = soup.find("table", class_="table")

# Find all the rows in the table
rows = table.tbody.find_all("tr")

# Create a list to store the player information
players = []

# Loop through each row in the table
for row in rows[:10]:
    # Find the columns in the row
    columns = row.find_all("td")

    # Extract the player information from the columns
    rank = int(columns[0].text)
    name = columns[1].text.strip()
    team = columns[2].text.strip()
    rating = int(columns[3].text)

    # Add the player information to the list
    players.append({
        "Rank": rank,
        "Name": name,
        "Team": team,
        "Rating": rating
    })

# Create a pandas dataframe from the list of players
df = pd.DataFrame(players)

# Print the dataframe
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"


response = requests.get(url)


soup = BeautifulSoup(response.content, "html.parser")


section = soup.find("section", class_="browse-list-section")


articles = section.find_all("li", class_="browse-list-item")


data = []


for article in articles:

    title = article.find("h3", class_="title").text
    authors = article.find("div", class_="authors").text
    published_date = article.find("div", class_="published-date").text


    link = article.find("a", class_="title")["href"]
    url = f"https://www.journals.elsevier.com{link}"


    data.append({
        "Title": title,
        "Authors": authors,
        "Published Date": published_date,
        "URL": url
    })


df = pd.DataFrame(data)


df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# request page data
page = requests.get("https://www.dineout.co.in/delhi-ncr-restaurants")
soup = BeautifulSoup(page.content, "html.parser")

```

```python
# create empty lists to store data
rest_name = []
cuisine = []
location = []
ratings = []
img_url = []

# loop through the page to find restaurant details
for restaurant in soup.find_all("div", class_="rest-info-wrap"):
    # get restaurant name
    rest_name.append(restaurant.find("h3").text.strip())
    
    # get cuisine
    cuisine.append(restaurant.find("p", class_="category").text.strip())
    
    # get location
    location.append(restaurant.find("p", class_="location").text.strip())
    
    # get ratings
    ratings.append(restaurant.find("p", class_="rating").text.strip())
    
    # get image url
    img = restaurant.find("img")
    img_url.append(img.get("src"))

# create data frame from collected data
df = pd.DataFrame({"Restaurant Name": rest_name, "Cuisine": cuisine, "Location": location, "Ratings": ratings, "Image URL": img_url})
df
```
